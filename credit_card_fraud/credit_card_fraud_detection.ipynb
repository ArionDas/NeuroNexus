### Column : dob

#### The complete date of birth is not important, so let's just extract the age of the customer from it

```python
train_data_['dob'].unique()
```

#### So, the data format is uniform

```python
# Defining a function to extract the age of customer from their date of birth
def find_age(date_of_birth):
    year = int(date_of_birth[:4])
    return (2023 - year)

# Applying the function to the required column
train_data_['dob'] = train_data_['dob'].apply(find_age)

train_data_.rename(columns={'dob':'age'}, inplace=True)
train_data_['age'].unique()
```

### Column : merchant

```python
train_data_['merchant'].unique()
```

### Column : category

```python
train_data_['category'].unique()
```

### Columns : first, last - Full name can be found by joining them

```python
train_data_['name'] = train_data_['first'] + ' ' + train_data_['last'].astype(str)
train_data_['name'].unique()
train_data_ = train_data_.drop(['first'], axis=1)
train_data_ = train_data_.drop(['last'], axis=1)
train_data_.head()
```

### Column : Street
#### We have already been provided with the longitude and latitude of the customer's place, so we can drop this

```python
train_data_ = train_data_.drop(['street'], axis=1)
train_data_.sample(7)
```

### Column : trans_date_trans_time

```python
# This column is written as strings, I will parse it to date time format
train_data_['trans_date_trans_time'] = pd.to_datetime(train_data_['trans_date_trans_time'])
train_data_['trans_date_trans_time'].unique()
```

```python
# Extracting years, months and dates from this column
train_data_['year'] = train_data_['trans_date_trans_time'].dt.year
train_data_['month'] = train_data_['trans_date_trans_time'].dt.month
train_data_['day'] = train_data_['trans_date_trans_time'].dt.day

train_data_.drop('trans_date_trans_time', axis=1, inplace=True)
train_data_.sample(5)
```

```python
# Checking out the categorical features once again
column_names = train_data_.columns

categorical_columns = [var for var in column_names if train_data_[var].dtype=='O']

print("The columns with categorical data are: {}".format(categorical_columns))
```

```python
# It's safe to drop the 'name' column as it doesn't help in differentiating between customers
categorical_columns = categorical_columns[:-1]
categorical_columns
```

## Numerical Data

### The distance between the merchant and the customer may well be a factor, so we use the latitudes and longitudes to find the distance between them

```python
train_data_['latitude difference'] = abs(train_data_['lat']-train_data_['merch_lat'])
train_data_['longitude difference'] = abs(train_data_['long']-train_data_['merch_long'])
```

### Now, we have to find the displacement between the longitude difference and the latitude difference
### We can use Pythagorean Theorem for finding this value

```python
# Difference b/w. consecutive longitudes and latitudes is 110km
train_data_['displacement'] = np.sqrt(pow((train_data_['latitude difference']*110),2) + pow((train_data_['longitude difference']*110),2))
train_data_.sample(5)
```

```python
train_data_['displacement'] = round(train_data_['displacement'])
train_data_ = train_data_.drop(columns=['lat','long','merch_lat','merch_long','latitude difference','longitude difference'], axis=1)
train_data_.head()
```

```python
## dropping the city,state and zip columns as now we have the distance required
train_data_ = train_data_.drop(['city','state','zip'], axis=1)
train_data_.head()
```

```python
## dropping the Unnamed column as it is repetition
train_data_ = train_data_.drop(['Unnamed: 0'], axis=1)
train_data_.head()
```

## UNIX TIME & CC_NUM
### The number of seconds passed from the UNIX EPOCH i.e. 00:00:00 UTC on 1 January 1970
### These two features will help us in differentiating between customers

### So we can drop the name column which doesn't really contribute much

```python
train_data_ = train_data_.drop(['name'], axis=1)
train_data_.head()
```

```python
### Now, we can categorize the displacement column
train_data_['displacement'].unique()
```

```python
train_data_['displacement'].describe()
```

### Now, we can categorize the displacement column into 'near', 'far', 'very far' using the lower quartile and the upper quartile

```python
train_data_.loc[(train_data_['displacement']<62),['proximity']] = "near"
train_data_.loc[((train_data_['displacement']>=62) & (train_data_['displacement']<108)), ['proximity']] = "far"
train_data_.loc[(train_data_['displacement']>=108), ['proximity']] = "very far"

train_data_.sample(10)
```

### Similarly, categorizing the city population as "less dense","normal","over-crowded"

```python
train_data_['city_pop'].describe()
```

```python
train_data_.loc[(train_data_['city_pop']<10000), ['city_population_status']] = "less dense"
train_data_.loc[((train_data_['city_pop']>1000) & (train_data_['city_pop']<50000)),['city_population_status']] = "normal"
train_data_.loc[(train_data_['city_pop']>50000),['city_population_status']] = "over crowded"

train_data_.sample(5)
```

```python
train_data_ = train_data_.drop(['city_pop'], axis=1)
```
